# TEST API

```python
import requests
import json
import pandas as pd
```

```python
LOCAL_HOST = 'http://localhost:8000'
PREDICTION_ENDPOINT = '/predict/'
```

## PREDICT

```python
input_data = {
  "purchase_lead": 80,
  "length_of_stay": 1,
  "flight_hour": 23,
  "flight_day": 1,
  "route": "AKLKUL",
  "booking_origin": "Malaysia",
  "wants_extra_baggage": 1,
  "wants_preferred_seat": 1,
  "wants_in_flight_meals": 1,
  "flight_duration": 1,
  "num_passengers": 1,
  "sales_channel": "Internet",
  "trip_type": "RoundTrip"
}
```

```python
predictions = requests.post(LOCAL_HOST + PREDICTION_ENDPOINT,
                           json = input_data)
```

```python
predictions.json()
```

```python
predictions.json()['booking_prediction']
```
